```python
table["vsm_enero"] = table["vsm_actual_y"]
table["vsm_febrero"] = table["vsm_actual_x"]
table["PagosVencidos_enero"] = table["PagosVencidosMeses_y"]
table["PagosVencidos_febrero"] = table["PagosVencidosMeses_x"]
table["RemainingDebt_enero"] = table["SaldoInsolutoActual_y"]
table["RemainingDebt_febrero"] = table["SaldoInsolutoActual_x"]

Table = table[["key", "vsm_enero", "vsm_febrero", "PagosVencidos_enero", "PagosVencidos_febrero", "RemainingDebt_enero", "RemainingDebt_febrero", "estado"]]
Table.head()
```

```python
Table["saldo_pesosE"] = Table["vsm_enero"]*Table["RemainingDebt_enero"]
Table["saldo_pesosF"] = Table["vsm_febrero"]*Table["RemainingDebt_febrero"]
Table["DueDaysE"] = Table["PagosVencidos_enero"]*30
Table["DueDaysF"] = Table["PagosVencidos_febrero"]*30
Table.head()
```

```python
Performance_Stay = Table.loc[Table["DueDaysE"] == Table["DueDaysF"] , :]
Performance_Stay["Performance"] = "stay"
Performance_Stay.head()
#Performance_Stay.count()
```

```python
Performance_Rol = Table.loc[Table["DueDaysE"] < Table["DueDaysF"] , :]
Performance_Rol["Performance"] = "Rol"
Performance_Rol.head()
```

```python
Performance_Cure = Table.loc[Table["DueDaysE"] > Table["DueDaysF"] , :]
Performance_Cure["Performance"] = "Cure"
Performance_Cure.head()
```

```python
#cargar la CNBV nivel nacional 36 meses
CV_cnbv=pd.read_csv('cartera_comparaCNBV1.csv')
CV_cnbv.head()
del CV_cnbv['institucion'] #ajuste pro problema es csv
#CV_cnbv['Institucion']=CV_cnbv[str('institucion')]
CV_cnbv.dtypes
#CV_cnbv['institucion'].astype(str)
CV_cnbv.head()
```

```python
#cargar base de desempleo
unemploy_rate=pd.read_csv('rate_nal_desempleo16a19.csv')
unemploy_rate.head()   
unemploy_rate.rename(columns={'Mes':'periodo'}, inplace=True)
unemploy_rate.head()
```

```python
#caclulate IMOR (CNBV define IMOR as delinquency rate (90+))
CV_cnbv.describe()
CV_cnbv['IMOR']=round((CV_cnbv['sdo_vencido']/CV_cnbv['sdo_total'])*100,2)
CV_cnbv.head()
#CV_cnbv['institucion'].astype('str')
#CV_cnbv['institucion'] = CV_cnbv['institucion'].astype('|S')
CV_cnbv['institucion']= np.where(CV_cnbv['ID']==1,'BBVA', 'BM')
CV_cnbv.head()
#CV_cnbv.dtypes
```

```python
#Separate and join Banking and BBVA just IMOR 
#CV_cnbv.groupby('institucion').count()
CV_banking=CV_cnbv.loc[CV_cnbv['institucion']=='BM',['periodo','IMOR']]
CV_banking
CV_BK1=CV_cnbv.loc[CV_cnbv['institucion']=='BBVA',['periodo','IMOR']]
CV_BK1
CV_join=pd.merge(CV_banking,CV_BK1,on='periodo', how='inner')
CV_join.rename(columns={'IMOR_x':'IMOR_BM','IMOR_y':'IMOR_BK1'}, inplace=True)
CV_join.head()
#CV_join.dtypes
```

```python
#merge with unemployment rates
CV_unemploy_join=pd.merge(CV_join,unemploy_rate,on='periodo', how='inner')
CV_unemploy_join.reset_index()
CV_unemploy_join.columns = CV_unemploy_join.columns.to_series().apply(lambda x: x.strip())
CV_unemploy_join.head()
#CV_unemploy_join.dtypes
```

```python
#Obtain correlation TEST1:  Unemployment rate vs. IMOR BK1
#from scipy import stats
#pearson_coef
test1=CV_unemploy_join[['tasa_desempleo','IMOR_BK1']]
correlation = test1.corr(method='pearson')
correlation
print(f'The correlation between Delinquency Rate of Bank1 vs. National Unemployment rate is:' 
      + str(correlation))
```

```python
# plot correlation
plt.title ('Correlation: Unemployment Rate vs Delinquency Rate of Bank1 - Jun16-May19')
plt.scatter(CV_unemploy_join['tasa_desempleo'],CV_unemploy_join['IMOR_BK1'],color='b', alpha=0.5)
plt.xlabel('Unemployment Rate %')
plt.ylabel('IMOR BK1 %')
#plt.show()
```

```python
# T test  based on test 1
#Chi-square test of independence of variables in a contingency table.
#H0 - There is no linear relationship between Delinquency Rate in Bank1 and National Umemployment Rate from June 2016 to May 2019
#H1 - There is a linear relationship between Delinquency Rate in Bank1 and National Umemployment Rate from June 2016 to May 2019

stats.ttest_ind(CV_unemploy_join['IMOR_BK1'],CV_unemploy_join['tasa_desempleo'], equal_var = False)
#conclusion:  can not reject H0
```

```python
#Non Parametric test- Compute the Mann-Whitney rank test on samples x and y.
#Use only when the number of observation in each sample is > 20 and you have 2 independent samples of ranks
#they are ranked by time
import scipy
scipy.stats.mannwhitneyu(CV_unemploy_join['tasa_desempleo'],CV_unemploy_join['IMOR_BK1'], use_continuity=True, alternative=None)
#conclusion: can not reject H0
#CV_unemploy_join.to_csv('CV_unemploy_join.csv')
```
